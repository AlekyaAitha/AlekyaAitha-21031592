```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("diabetes.csv")
df['Outcome']=df['Outcome'].replace([1,0],['Diabetes','Healthy'])
df.head()
```

```python
def areaplot(fet,w,h):    # function to plot Area chart of the feature
    """ This function will plot the area chart by crosss tabuation 
        with the target label that is 'Outcome' Parameters will be taken 
        from the user to visualiae the area plot concerning that feature"""
    pd.crosstab(df[fet],df['Outcome']).plot(kind='area',figsize=(w,h),color=["#FF0000","#008000"])
    plt.title("Diabetes Disease by {}".format(fet),fontsize=18,color="r")
    plt.xlabel("{}".format(fet),fontsize=13,color="m")
    plt.ylabel("Count",fontsize=13,color="m")
    plt.grid()
    plt.show()
```

```python
def piechart(fet,dim):   # funstion to visualize the pie chart
    """ This function will plot the pie chart target label that 
        is 'Outcome' for the average value of the feature Parameters 
        will be taken from the user to visualiae the area plot 
        concerning that feature"""
    dib=df[df['Outcome']=="Diabetes"]
    hel=df[df['Outcome']=="Healthy"]
    gls=[dib[fet].mean(),hel[fet].mean()]
    plt.figure(figsize=(dim,dim))
    plt.pie(gls,labels=df['Outcome'].unique(),autopct='%1.2f%%',colors=["#FF0000","#008000"])
    plt.title("{} Levels".format(fet),fontsize=18,color="r")
    plt.show()
```

```python
def violinchart(fet):     # funstion to visualize the pie chart
    """ This function will plot the violon chart target label that 
        is 'Outcome' for the value of the feature Parameters 
        will be taken from the user to visualiae the area plot 
        concerning that feature"""
    ax = sns.violinplot(x=fet, y="Outcome", data=df,palette="rainbow")
    plt.title("{} Levels".format(fet),fontsize=18,color="r")
    plt.xlabel("{}".format(fet),fontsize=13,color="m")
    plt.ylabel("Outcome",fontsize=13,color="m")
    plt.grid()
    plt.show()
```

```python
areaplot('Age',7,4)   # call areaplot fucntion
```

```python
piechart("Glucose",7)    # call piechart fucntion
```

```python
piechart("BloodPressure",7)      # call piechart fucntion
```

```python
violinchart("Insulin")       # call violin lot fucntion
```
